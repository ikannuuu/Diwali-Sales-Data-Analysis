```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape #To get how many rows and columns in data set
```

```python
df.head() #To see the first five rows of the dataset
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace = True)
#Drop funtion for deleting the unwanted values where axis is for specifying the whole column and inplace in use to save the change after deletion
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
#for checking the no. of null values in each column
```

```python
df.shape
```

```python
df.dropna(inplace=True) #Drop the null values
```

```python
df.shape
```

```python
pd.isnull(df).sum()
#for checking the no. of null values in each column
```

```python
df['Amount']= df['Amount'].astype('int')
#Change data type
```

```python
df['Amount'].dtype
#check the type of the column
```

```python
df.columns
#Checking what columns are present in the dataset
```

```python
df.rename(columns = {'Marital_Status':'Married'})
#rename the columns
```

```python
df.describe()
#Describe fun. is used to get the description of the data (i.e mean, count, std, etc)
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df )
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
Sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = Sales_gen)
plt.show()
```

**From above graphs we can see that most of the buyers are females and even the purchasing power of the females are greater than men**

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
Sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = Sales_age)
plt.show()
```

**From the above graph we can see most of the buyers are from age group 26-35 years females**

```python
df.columns
```

```python
#Total numbers of orders from top 10 states
Sales_age = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(16,6)})
sns.barplot(x = 'State', y = 'Orders', data = Sales_age)
plt.show()
```

```python
#Total sales/ top 10
Sales_age = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(16,6)})
sns.barplot(x = 'State', y = 'Amount', data = Sales_age)
plt.show()
```

**From the above graph we can see that most of the orders are from utter pradesh, maharashtra and karnataka respectively**

```python
df.columns
```

```python
ax = sns.countplot(data = df, x='Marital_Status')

sns.set(rc = {'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
Sales_Marital_Status = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(10,6)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = Sales_Marital_Status, hue = 'Gender')
plt.show()
```

**From the above graph we can see that most of the buyers are married women and they have high purchasing power**

```python
sns.set(rc={'figure.figsize':(20,6)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
Sales_Occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(x = 'Occupation', y = 'Amount', data = Sales_Occupation)
plt.show()
```

**From the above graph we can see that most of the buyers are from IT, Healthcare and Aviation sector**

```python
sns.set(rc={'figure.figsize':(25,6)})
ax = sns.countplot(data = df, x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
Sales_Product_Category = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(28,6)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = Sales_Product_Category)
plt.show()
```

**From the above graph we can see that most of the sold product are from Food, Clothing & apprarel and Electronics & Gadgets**

```python
Sales_Product_ID = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(28,6)})
sns.barplot(x = 'Product_ID', y = 'Orders', data = Sales_Product_ID)
plt.show()
```

**CONCLUSION:**
*Married women age group 26-35 years from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely buy products from food, Clothing & apprarel and Electronics & Gadgets Category*
